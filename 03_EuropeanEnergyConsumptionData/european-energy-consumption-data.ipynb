 We can select other kind of graphic for display the information. Shall we select a bar chart? The information about the bar chart is üëá

```python
axes = df.plot.bar(rot=90, subplots=True)
axes[1].legend(loc=2) 
```

ü§î Shall we improve? We are going to display an horizontal bar plot which show the percentage over the total energy consumption in Europe per country. But first we need to calculate the total energy consumption in Europe.

```python
##Calculate percentage of energy source

# Calculate the sum of each column.
df_general = df.sum(axis=0)

# Quotient of each country by the sum of all countries, multiplican by 100.
df_percentage = (df/df_general.sum())*100

# Sorting energy values ascending
df_percentage.sort_values(by='energy', ascending=True, inplace=True)

# Display de horizontal bar plot
df_percentage['energy'].plot.barh()
```

The result is consistent with the result of maximun and minimun energy values, is not it?

üìå Are energy and gdp correlated? According with the scatter plot üëá ... could be!

```python
df.plot.scatter(x='energy', y='gdp')
```

```python
# Pearson's r
print( 'Pearson:', df['energy'].corr(df['gdp']))
# Spearman's rho
print( 'Spearman:', df['energy'].corr(df['gdp'], method='spearman') )
```

The pearson coeficient near 1 tell us the strong correlation between the energy and the gdp.
