**About Dataset:** 

9 columns:
- **Year**: Year the contest took place.
- **Country**: Which country won the contest.
- **Song**: Title of the song which won the contest
- **Artist**: Name of the participant(s) who won the contest
- **Songwriter(s)**: People or persons who wrote the winning song.
- **Running Order**: Where the winning song was placed in the running order of the Grand Final.
- **Semi Final Points**: Points received in the semi-final (if the winner took part).
- **Total Points**: Total points received in the Grand Final.
- **Location**: City in which the contest was hosted.

## Display the country which won more times Eurovision

```python
#count the number of times one country is in 'Country' columns

#Define a serie serie_country , selecting only the 'Country' column ohe bracket
serie_country = df['Country']
#It is not a DataFrame is a serie
print('Type {0}'.format(type(serie_country)))

# and count the same value appears in the column
count_country = serie_country.value_counts()

#print the 10th first rows in count_country
print(count_country.head(10))
```

```python
#print df_country DataFrame before dropping one row
print(' Shape {0}, before dropping one row'.format(count_country.shape))

#Remove the row with the year the contest was cancelled, in the same dataframe
count_country.drop('Contest cancelled due to the COVID-19 pandemic', inplace=True)

#print df_country DataFrame after dropping one row
print(' Shape {0}, after dropping one row'.format(count_country.shape))
```

### Graph Visualization

```python
#Setting the plot size
fig, ax = plt.subplots(figsize=(15, 10))

#Plot type - bar with the values from count_country
ax = sns.barplot(
            # X number of times the country won
            x = count_country.values,
            # Y name of the country
            y = count_country.index,
            # set the color_graph as color
            color=color_graph
            )

# Display the number 
ax.bar_label(ax.containers[0])

#Adding title 
plt.title('Eurovision Song Contest Winners 1956 - 2022')

#Display plot
plt.show()
```

```python
#Selecting the maximun values of the count_country
max_won = count_country.max()

#Using the max_won value for selecting the country name
Country_max = count_country[count_country == max_won]

#Display nice and tidy the information using format
# The information about the name of the country is the first element of the index
print('{0} won Eurovision {1} times'.format(
    #first element printing - country name
    Country_max.index[0], 
    #second element printing - number of time this country won
    max_won ))
```

```python
#Selectining the column year before selecting the value in 'country' column
#The year without contest
df[df['Country'] == 'Contest cancelled due to the COVID-19 pandemic']['Year']
```

### the year with ? winners

```python
#Define a serie, selecting and counting different year
serie_year  = df['Year'].value_counts()

#selecting the year with more than one winner
year = serie_year[serie_year != 1].index[0]

df_year = df[df['Year']==year]

print('In {0}, {1} hosted the contest and it was {2} wins with {3} points each\n'.format(
    #first value printing - year
    year,    
    #second value printing - city hosted (Advance)
    df_year[df_year['Location'].notnull()]['Location'].values[0],
    #third value printing - number of rows = number of winners
    df_year.shape[0], 
    #third value printing -  unique value of the column 'Total Points'
    df_year['Total Points'].unique()[0]
))
```

##  What is the Points Trend?

```python
df_points = df
fig, ax = plt.subplots(figsize=(15, 10))
_ = sns.lineplot(x=df['Year'], 
                 y=df['Semi Final Points'], 
                 data=df,
                color=color_graph)
_ = sns.lineplot(x=df['Year'], 
                 y=df['Total Points'], 
                 data=df,
                color=color_background)

plt.title('Semi Final & Total Points Trend')
plt.show()
```

<a id=Correlation></a>
## There are 'Running Order', 'Semi Final Points', and 'Total Points'correlated 

```python
#Display a triangle heatmap

#Define data frame Selecting columns (two brackets)
df_corr = df[['Running Order',  'Semi Final Points', 'Total Points']]

# creating mask - Triangle heatmap
mask = np.triu(np.ones_like(df_corr.corr()))

fig, ax = plt.subplots(figsize=(15, 10))
_ = sns.heatmap(
                #data
                df_corr.corr(), 
                #triangle display
                mask=mask,
                #setting color_graph   
                cmap=sns.light_palette(color_graph , as_cmap=True),  
                #Try the following line
                #cmap=sns.light_palette('red', as_cmap=True),
                #display correlation number with two decimal places format
                annot=True, 
                fmt=".2f")
plt.show()
```

üßÆ slight correlation between  'Semi Final Points', and 'Total Points'
